```python
###This is to show how to read, write, and append text files!
```

```python
###location of main text folder:
mainLoc = "C:\\Users\\fabri\\OneDrive\\Documents\\DasText"
```

```python
##list of closer lyrics by chainsmokers
lyricsLoc = mainLoc+"\\closerLyrics.txt"
##list of random integers text
intsLoc = mainLoc+"\\random10kIntegers.txt"
##New file loc that I want to write
newLoc = mainLoc+"\\newGeneratedText.txt"
```

```python
##We can open a document in one of two way... with a simple open command, OR with a "with" statement
##simple open command:
lyricsFile = open(lyricsLoc,"r")
```

```python
##store the contents of lyricsFile into list lyrics
lyrics = lyricsFile.read()
```

```python
##make a list of every line in lyrics with .split("\n")
##for list of every word, just use .split() with no parameter or the " " parameter
everyLine = lyrics.split("\n")
```

```python
everyLine
```

```python
##However, if we use a simple open command, that file will remain "open"... it 
## is MUCH better practice to open with a "with" command!
## here we show having to close the file manually
print(lyricsFile.closed)
lyricsFile.close()
print(lyricsFile.closed)
```

```python
##using "with" statement, which automatically closes the file afterwards
with open(intsLoc,"r") as integerDataFile:
    integerList = integerDataFile.read()
    print(integerList)
```

```python
##we split the integer list into an actualy list of integers in the file
integerList = integerList.split("\n")
```

```python
##let's get the sum, but first we need to make it into int array!
integerList = [int(i) for i in integerList]
integerList
```

```python
sum(integerList)
```

```python
##now let's MAKE a document... use the module "w" to write!
with open(newLoc,"w") as newTextFile:
    newTextFile.write("This is a new text document that I started.\nI will be appending to it everytime I log on to document my stuff! jkjk\n")
```

```python
##you can open it on your computer to make sure, or just check here!
with open(newLoc,"r") as newTextFile:
    newTextData = newTextFile.read()
    print(newTextData)
```

```python
##Now let's add to our newly created document! use module "a" for append!
##we will also check it here again to make sure it was added.
with open(newLoc,"a") as newTextFile:
    newTextFile.write("\nTHIS IS AN APPENDED TEXT!\nWE CAN ADD TO IT AS MUCH AS WE WANT!\n")
with open(newLoc,"r") as newTextFile:
    newTextData = newTextFile.read()
    print(newTextData)
```
