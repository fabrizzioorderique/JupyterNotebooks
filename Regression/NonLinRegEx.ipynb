```python
####Taking a look at non-linear regression models 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
####LINEAR EXAMPLE

##making the x-coordinates
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3

#the dots
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

#plt.figure(figsize=(8,6))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
```

```python
####CUBIC EXAMPLE

x = np.arange(-5.0, 5.0, 0.1)
y = 1*(x**3) + 1*(x**2) + 1*x + 3
#dots
y_noise = 20*np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata,'bo')
plt.plot(x,y, 'r') 
```

```python
####QUARTIC

x = np.arange(-5.0, 5.0, 0.1)
y = np.power(x,2)

y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
```

```python
####Exponential
X = np.arange(-5.0, 5.0, 0.1)
Y= np.exp(X)

plt.plot(X,Y) 
```

```python
####LOGARITHMIC

X = np.arange(-5.0, 5.0, 0.1)
Y = np.log(X)
plt.plot(X,Y) 
```

```python
####Sigmoidal/Logistic

X = np.arange(-5.0, 5.0, 0.1)
Y = 1-4/(1+np.power(3, X-2))

y_noise = 0.2 * np.random.normal(size=x.size)
ydata = Y + y_noise

plt.plot(X,ydata,'bo')
plt.plot(X,Y,'r') 
```
