```python
##Welcome! We are going to be plotting and making some simple graphs today.
## I'll be using to example data sets: randomIntegers10k and AprilFoolsDeaths.csv,
    ## So it is reccomended that you check out how to read and write csv and txt files!

##let's start by importing necessary libraries:

import pandas as pd #for reading csv files
import numpy as np #for using math and vector manipulation
import matplotlib.pyplot as plt #for graphing!
```

```python
##let's make a sine graph!

## We need to use np.linspace() to generate num numbers evenly spcaed in an interval and np.pi to get pi value
x = np.linspace(0,4*np.pi,num = 15)
print("x values: ")
print(x)
y = np.sin(x)
print("\ny values: ")
print(y)
```

```python
#now let's plot these guys 
## we need to use %matplotlib inline AND plt.plot()
%matplotlib inline
plt.plot(x,y)
```

```python
##looks rough? yeah I think so too... let's add more x values then
x = np.linspace(0,4*np.pi,num = 100)
y = np.sin(x)
plt.plot(x,y)
```

```python
##Cool, let's try to make a cooler graph
x = np.linspace(-4,4,num=200)
y = (x*np.cos(x))/(np.tan(x**2))
plt.plot(x,y)
```

```python
##now let's make a histogram frequency table!
## let's define a sequence... number of hours I studied data science in the last 15 days
hours = [0,2,1,3,0,5,4,3,4,2,6,5,7,2,3]

##making the histogram
hist, bin_edges = np.histogram(hours)
print(hist)
print(bin_edges)
```

```python
##let's plot hours
n, bins, patches = plt.hist(x=hours, bins='auto', color='#0504aa',alpha=0.6, rwidth=0.85)
```

```python
n, bins, patches = plt.hist(x=hours, bins='auto', color='#0504aa', alpha=0.9, rwidth=0.85)
##Needs some labels and pizzaz... 
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram')
#plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
plt.ylim(ymax = 6)
```

```python
##Great! But now how do we plot a dataframe? say one with categorical vs numerical data

## maybe something like number of deaths in each state on AprilFools?
## see where I am going with this? --> let's read AprilFoolsDeaths.csv file that we created in csvManipulation!
statesDeathsDf = pd.read_csv("C:\\Users\\fabri\\OneDrive\\Documents\\DasText\\csvFiles\\AprilFoolsDeaths.csv")
statesDeathsDf.head()
```

```python
#ew let me get rid of first column:
statesDeathsDf = statesDeathsDf[['state','death']]
print(statesDeathsDf)
```

```python
#much better... but now how do we plot this? let's use PANDAS
pd.DataFrame(statesDeathsDf).plot(kind='bar',figsize=(12,4),title="Death Count by US State or Territory")
```

```python
#cool... but I want the state abbreviations to show up on the bottom :/ ...USING MATPLOT 
plt.figure(figsize=(20,4))
# Add title and axis names
plt.title('Death Count by US State or Territory')
plt.xlabel('State/Territory Abbreviation')
plt.ylabel('Deaths')
#now we plot it
plt.bar(statesDeathsDf['state'], statesDeathsDf['death'])
```

```python
#WOW! Look at NY! Let's compare the NY deaths to GA deaths
NyDeaths = statesDeathsDf.loc[statesDeathsDf['state']=="NY"]
print(NyDeaths)
GaDeaths = statesDeathsDf.loc[statesDeathsDf['state']=="GA"]
print(GaDeaths)
```
